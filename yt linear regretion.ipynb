```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("canada_per_capita_income.csv")
df
```

```python
print(df.columns)
```

```python
%matplotlib inline
plt.xlabel('year ' )
plt.ylabel('price')
plt.scatter(df.year , df.price , color = 'red' , marker='+')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['year']], df.price)
```

```python
# simplest — list of list (1 sample, 1 feature)
reg.predict([[1976]])
```

```python
reg.coef_
```
